```python
#Import necessary libraries
import pandas as pd
import numpy as np
import os
import time
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, roc_curve, auc, plot_confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

sns.set_style('darkgrid')

#Load in the hotel dataset from Kaggle as a csv file.
hotelData = pd.read_csv('Hotel Reservations.csv')
```

# Data Understanding
* Booking_ID - ID for each booking, this is not relevant and will be dropped later.
* no_of_adults - Number of adults per booking.
* no_of_children - Number of children per booking.
* no_of_weekend_nights - Number of weekend nights per booking.
* type_of_meal_plan - Meal plan per booking.
* required_car_parking_space - Preferred parking space.
* room_type_reserved - Type of room per booking.
* lead_time - Days booked in advance of the reservation.
* arrival_year - Year of booking.
* arrival_month - Month of booking.
* arrival_date - Date of booking.
* market_segment_type - Online or offline booking.
* repeated_guest - Customer booked at least one room previously.
* no_of_previous_cancellations - Number of cancellations previously.
* no_of_previous_bookings_not_canceled - Number of bookings that were not canceled.
* avg_price_per_room - Average price per room in dollars.
* no_of_special_requests - Number of special requests.
* booking_status - 0 if the booking was not canceled and 1 if the booking was canceled. This is the target variable and of particular interest.

```python
#Transforming object type variables.
le = LabelEncoder()
scaler = StandardScaler()

hotelData['room_type_reserved'] = le.fit_transform(hotelData['room_type_reserved'])
hotelData['type_of_meal_plan'] = le.fit_transform(hotelData['room_type_reserved'])
hotelData['booking_status'] = le.fit_transform(hotelData['booking_status'])
hotelData['market_segment_type'] = le.fit_transform(hotelData['market_segment_type'])

```

```python
#Dropping the Booking_ID column as it's not useful.
hotelData = hotelData.drop('Booking_ID', axis = 1)
hotelData
```

```python
#This horizontal bar graph shows the vast majority of customers are NOT repeat.
repeat_data = hotelData['repeated_guest'].value_counts()

#Plot labels and graphing information
bars_labels = 'Unique Customers', 'Repeat Customers'
plt.barh(bars_labels, repeat_data, color='blue')
plt.title('Repeat vs. Unique Customers')
plt.xlabel('Number of Customers')
plt.show()

#Print a message showing the absurdly low number of repeat customers.
average_repeat = repeat_data[1]/repeat_data[0]
print("Only",round(average_repeat*100, 2), "% are repeat customers.")

#0 not repeat
#1 repeat guest
```

```python
#Creating the heatmap to show correlation of variables.

correlation = hotelData.corr()
plt.figure(figsize=(18,8))
sns.heatmap(correlation, vmin=-1, vmax=1, center=0, annot= True)
plt.title('Heatmap')
```

```python
#I get the total cancelled and uncancelled reservations here then convert to list.
reservation_comparison = hotelData['booking_status'].value_counts()
values_for_plotting = list(reservation_comparison)

#This separates the cancelled vs uncancelled reservations.
notCancelled = values_for_plotting[0]
cancelled = values_for_plotting[1]

#Plotting the data with labels.
bar_labels = "Not Cancelled", "Cancelled"
plt.barh(bar_labels, reservation_comparison, color='red')
plt.title("Cancelled vs. Uncancelled Hotel Reservations")
plt.xlabel("Number of Reservations")
plt.show()

#This visualization shows the class imbalance between the reservation choices.
```

#    Data Preparation

```python
X = hotelData.drop('booking_status', axis=1).to_numpy()
y = hotelData['booking_status'].to_numpy()

#X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, 
                                                   random_state = 10)
```

```python
#Random Forest here

rfc = RandomForestClassifier(class_weight='balanced')
rfc.fit(X_train, y_train)
```

```python
y_preds_baseline = rfc.predict(X_test)
y_train_preds = rfc.predict(X_train)
```

```python
#build a confusion matrix and classification report

display(confusion_matrix(y_test, y_preds_baseline))
print('Classification Report of Random Forest: \nTrain')
print(classification_report(y_train, y_train_preds))
print('------------------------------------------------------ \nTest')
print(classification_report(y_test, y_preds_baseline))
```

```python
#Here's the initial decision tree
#I intentionally included a random_state=10 because I wanted reproducibility.

decision_tree_model = DecisionTreeClassifier(random_state=10)
decision_tree_model.fit(X_train, y_train)

#This is for predicting for the test set.
y_pred = decision_tree_model.predict(X_test)
```

```python
#Confusion matrix showing TP, FP, FN, TN
sns.heatmap(metrics.confusion_matrix(y_pred, y_test), annot=True, fmt='g')
# sns.heatmap(metrics.confusion_matrix(decision_tree_model, X, y), annot=True)


plot_confusion_matrix(decision_tree_model, X, y)
plt.show()
#True Positive, False Positive
#False Negative, True Negative
#Looking at the confusion matrix, we can see the vast majority 
#of the true positive and true negative values were accurately predicted.
```

```python
##Model Tuning via GridSearchCV
```

```python
#Create the parameter grid

param_grid = {'criterion':['gini', 'entropy'],
             'max_depth':[3,6,9,12],
             'min_samples_leaf':[5,10]
             }
```

```python
#Fit the gridsearch

rf = RandomForestClassifier(class_weight= 'balanced')
grid_rfc = GridSearchCV(rf, param_grid, cv = 5, scoring='accuracy')
grid_rfc.fit(X_train, y_train)
```

```python
#Print the best parameters

grid_rfc.best_params_
```

```python
#Get predictions from the gridsearch

y_preds_grid = grid_rfc.predict(X_test)
```

```python
#Print confusion matrix and classification report

display(confusion_matrix(y_test, y_preds_grid))
print(classification_report(y_test, y_preds_grid))
```
